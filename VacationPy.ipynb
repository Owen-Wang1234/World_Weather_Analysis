```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
import os

# Import the API key.
from config import g_key
```

```python
# Store the CSV saved from the WeatherPy program into a DataFrame.
city_data_file = os.path.join("weather_data", "cities.csv")
city_data_df = pd.read_csv(city_data_file)
city_data_df.head()
```

```python
# Verify the data types.
city_data_df.dtypes
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key = g_key)
```

```python
# Set up a heatmap for temperature.
# 1. Set up the latitude and longitude pairs.
locations = city_data_df[["Lat", "Lng"]]

# 2. Set up the weights for temperatures.
max_temp = city_data_df["Max Temp"]
temperatures = [max(temp, 0) for temp in max_temp]

# 3. Assign the figure variable to the gmaps.figure() attribute. Center and adjust zoom level.
fig = gmaps.figure(center = (30.0, 31.0), zoom_level = 1.5)

# 4. Assign the heatmap_layer variable to the heatmap_layer attribute and add in the locations.
heatmap_layer = gmaps.heatmap_layer(locations, weights = temperatures, dissipating = False,
                                    max_intensity = 300, point_radius = 4)

# 5. Add the heatmap layer.
fig.add_layer(heatmap_layer)

# 6. Call the figure to plot the data.
fig
```

```python
# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
```

```python
# Filter the dataset to find the cities that fit the criteria.
preferred_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) &
                                       (city_data_df["Max Temp"] >= min_temp)]
preferred_cities_df.head(10)
```

```python
# Filter out any full values and see if enough results are produced.
preferred_cities_df = preferred_cities_df.dropna()
preferred_cities_df.count()
```

```python
# Create a hotel DataFrame to store hotel names along with city, country, max temp, and coordinates.
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Set up the URL and the parameters.
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

params = {"radius": 5000,
          "type": "lodging",
          "key": g_key}

# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and the longitude.
    lat = row["Lat"]
    lng = row["Lng"]
    
    # Add the latitude and longitude to the location key for the params dictionary.
    params["location"] = f"{lat},{lng}"
    
    # Make the request and get the JSON data from the search.
    hotels = requests.get(url, params = params).json()
    
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    
    except (IndexError):
        print("Hotel not found... skipping.")

hotel_df.head(10)
```

```python
# Prepare the hotel information for markers.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} \N{DEGREE SIGN}F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
hotel_info
```

```python
# Heatmap of temperature for the vacation spots
# Set up the coordinates and weights.
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]

# Set up the figure with preset center and zoom.
fig = gmaps.figure(center = (30.0, 31.0), zoom_level = 1.5)

# Set up the heatmap layer.
heatmap_layer = gmaps.heatmap_layer(locations, weights = max_temp, dissipating = False,
                                    max_intensity = 300, point_radius = 4)

# Add the marker layer.
marker_layer = gmaps.marker_layer(locations, info_box_content = hotel_info)

# Assemble and display the figure.
fig.add_layer(heatmap_layer)
fig.add_layer(marker_layer)
fig
```
